```python
from web3 import Web3
from solcx import compile_source, install_solc

# Install Solidity Compiler
install_solc("0.8.0")

# Ganache RPC URL from Ngrok
ganache_url = "https://6908-180-235-122-218.ngrok-free.app"
w3 = Web3(Web3.HTTPProvider(ganache_url))

# Check connection
if w3.is_connected():
    print("‚úÖ Connected to Ganache")
else:
    raise Exception("‚ùå Failed to connect")

# Replace with your Ganache Private Key
private_key = "0x7f9d0074ff2613555cdb8f17c7afbf8e65d7c4d91aaa3ab106a429dc840e25e9"
account = w3.eth.account.from_key(private_key)
w3.eth.default_account = account.address

# Solidity Smart Contract Source Code
contract_source_code = """
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract Storage {
    uint256 private number;

    function setNumber(uint256 _num) public {
        number = _num;
    }

    function getNumber() public view returns (uint256) {
        return number;
    }
}
"""

# Compile Contract
compiled_sol = compile_source(contract_source_code, solc_version="0.8.0")
contract_id, contract_interface = compiled_sol.popitem()

# Get Bytecode and ABI
bytecode = contract_interface["bin"]
abi = contract_interface["abi"]

# Deploy Contract
Storage = w3.eth.contract(abi=abi, bytecode=bytecode)
tx = Storage.constructor().build_transaction({
    "from": w3.eth.default_account,
    "nonce": w3.eth.get_transaction_count(w3.eth.default_account),
    "gas": 2000000,
    "gasPrice": w3.to_wei("10", "gwei")
})

# Sign and Send Transaction
signed_tx = w3.eth.account.sign_transaction(tx, private_key)
tx_hash = w3.eth.send_raw_transaction(signed_tx.raw_transaction)  # Corrected
tx_receipt = w3.eth.wait_for_transaction_receipt(tx_hash)

# Get Contract Address
contract_address = tx_receipt.contractAddress
print("‚úÖ Contract Deployed at:", contract_address)
```

```python
# Get the contract instance
contract = w3.eth.contract(address=contract_address, abi=abi)
```

```python
# Prepare transaction to set a number (e.g., 42)
tx = contract.functions.setNumber(42).build_transaction({
    "from": w3.eth.default_account,
    "nonce": w3.eth.get_transaction_count(w3.eth.default_account),
    "gas": 100000,
    "gasPrice": w3.to_wei("10", "gwei")
})

# Sign and send transaction
signed_tx = w3.eth.account.sign_transaction(tx, private_key)
tx_hash = w3.eth.send_raw_transaction(signed_tx.raw_transaction)

# Wait for transaction receipt
tx_receipt = w3.eth.wait_for_transaction_receipt(tx_hash)
print("‚úÖ Number set successfully!")
```

```python
# Call the getNumber function
stored_number = contract.functions.getNumber().call()
print("üìå Stored Number in Contract:", stored_number)
```
